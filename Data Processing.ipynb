```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
subfolders2 = os.listdir('/content/drive/MyDrive/datasets/human-body-segmentation dataset/masks/')
subfolders1 = os.listdir('/content/drive/MyDrive/datasets/human-body-segmentation dataset/images/')
```

```python
plt.imshow(plt.imread('/content/drive/MyDrive/datasets/human-body-segmentation dataset/images/'+ subfolders1[0]))
```

```python
plt.imshow(plt.imread('/content/drive/MyDrive/datasets/human-body-segmentation dataset/masks/'+ subfolders2[0]))
```

```python
images = []
for i in range(len(subfolders1)):
    filepath = '/content/drive/MyDrive/datasets/human-body-segmentation dataset/images/' + subfolders1[i]
    image = plt.imread(filepath)
    images.append(image)
```

```python
masks = []
for i in range(len(subfolders2)):
    filepath = '/content/drive/MyDrive/datasets/human-body-segmentation dataset/masks/' + subfolders2[i]
    image = plt.imread(filepath)
    masks.append(image[:, :, 0])
```

```python
masks_array = np.array(masks)
images_array = np.array(images)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(images_array, masks_array, test_size = 0.25)
```

```python
np.save('/content/drive/MyDrive/datasets/human-body-segmentation dataset/human_segmentation_train_masks.npy', y_train)
np.save('/content/drive/MyDrive/datasets/human-body-segmentation dataset/human_segmentation_train_images.npy', X_train)
np.save('/content/drive/MyDrive/datasets/human-body-segmentation dataset/human_segmentation_test_masks.npy', y_test)
np.save('/content/drive/MyDrive/datasets/human-body-segmentation dataset/human_segmentation_test_images.npy', X_test)
```
